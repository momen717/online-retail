# preprocessing

```python
countries=df1['Country'].value_counts()
countries
```

```python
df1['TotalAmount'] = df1['Quantity'] * df1['Price']
df1.head()
```

```python
df1['InvoiceDate'] = pd.to_datetime(df1['InvoiceDate'], format="%d-%m-%Y %H:%M")
last_date = max(df1['InvoiceDate'])
df1['TimeDifference'] = (last_date - df1['InvoiceDate']).dt.days
df1.head()
```

# RFM
### R (Recency)
### F (Frequency) 
### M (Monetary)

```python
recency = df1.groupby('Customer ID')['TimeDifference'].min()

customer1 = pd.DataFrame(recency, index=recency.index)
customer1.columns = ['Recency']
customer1
```

```python
customer1['Frequency'] = df1.groupby('Customer ID')['Invoice'].unique().apply(lambda x: len(x))
customer1.head()
```

```python
customer1['Monetary'] = df1.groupby('Customer ID')['TotalAmount'].sum()
customer1.head()
```

# Interpretation:
### High Recency, High Frequency, High Monetary: Likely your best customers.
### Low Recency, High Frequency, High Monetary: Loyal customers who haven't purchased recently.
### High Recency, Low Frequency, Low Monetary: New or one-time customers.
### Low Recency, Low Frequency, Low Monetary: Customers who might be at risk of churning.

# clustering 1

```python
model = KMeans(3)
customer1['cluster'] = model.fit_predict(customer1)

fig, ax = plt.subplots(1, 3, figsize=(14, 4))
sns.scatterplot(customer1, x='Recency', y='Frequency', ax=ax[0], hue='cluster', palette='bright')
sns.scatterplot(customer1, x='Recency', y='Monetary', ax=ax[1], hue='cluster', palette='bright')
sns.scatterplot(customer1, x='Frequency', y='Monetary', ax=ax[2], hue='cluster', palette='bright')
```

```python
scaler = StandardScaler()
result = scaler.fit_transform(customer1.drop('cluster', axis=1))
scaled_df = pd.DataFrame(result)
scaled_df.columns = customer1.columns[:-1]
scaled_df.head()
```

```python
model = KMeans(3)
scaled_df['cluster'] = model.fit_predict(scaled_df)

fig, ax = plt.subplots(1, 3, figsize=(12, 4))
sns.scatterplot(scaled_df, x='Recency', y='Frequency', ax=ax[0], hue='cluster', palette='bright')
sns.scatterplot(scaled_df, x='Recency', y='Monetary', ax=ax[1], hue='cluster', palette='bright')
sns.scatterplot(scaled_df, x='Frequency', y='Monetary', ax=ax[2], hue='cluster', palette='bright')
```

```python
silhouette_score(scaled_df, scaled_df['cluster'])
```

```python
values = {'k': [], 'inertia': []}

for k in range(2, 20):
    model = KMeans(k)
    model.fit(scaled_df)
    values['k'].append(k)
    values['inertia'].append(model.inertia_)
```

```python
sns.lineplot(x=values['k'], y=values['inertia'], marker='o')
```

```python
model = KMeans(9)
scaled_df['cluster'] = model.fit_predict(scaled_df)

fig, ax = plt.subplots(1, 3, figsize=(12, 4))
sns.scatterplot(scaled_df, x='Recency', y='Frequency', ax=ax[0], hue='cluster', palette='bright')
sns.scatterplot(scaled_df, x='Recency', y='Monetary', ax=ax[1], hue='cluster', palette='bright')
sns.scatterplot(scaled_df, x='Frequency', y='Monetary', ax=ax[2], hue='cluster', palette='bright')
```
